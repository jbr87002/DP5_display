```python
import os

imgs_folder_path = os.path.join(calc_folder_path, 'imgs')
svg_contents = []
for filename in os.listdir(imgs_folder_path):
    if filename.endswith('.svg'):
        file_path = os.path.join(imgs_folder_path, filename)
        with open(file_path, 'r') as f:
            svg_content = f.read()
            svg_contents.append(svg_content)


for svg_content in svg_contents:
    display(SVG(svg_content))
```
